# Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
pd.set_option("display.max_columns", None)
```

##### Successfully imported all required libraries for data analysis and visualization.

# Step 2: Load Excel Dataset

```python
file_path="Global_Covid19_Data.xlsx"
covid_data=pd.read_excel(file_path,sheet_name="covid_19_data")
country_daywise=pd.read_excel(file_path,sheet_name="country_daywise")
countrywise=pd.read_excel(file_path,sheet_name="countrywise")
country=pd.read_excel(file_path,sheet_name="Country")
vaccination=pd.read_excel(file_path,sheet_name="Vaccination")
vaccination_detail=pd.read_excel(file_path,sheet_name="Vaccinations_detail")
daywise=pd.read_excel(file_path,sheet_name="daywise")
```

##### Successfully loaded all Excel worksheets into separate DataFrames.

# Step 3: Explore Dataset

```python
datasets={"covid_data":covid_data,"country_daywise":country_daywise,"countrywise":countrywise,"country":country,"vaccination":vaccination,"vaccination_detail":vaccination_detail,"daywise":daywise}
for name,df in datasets.items():
    print(name)
    display(df.head())
```

##### Displayed sample records to understand the structure of each dataset.

# Step 4: Check Dataset Shape

```python
for name,df in datasets.items():
    print(name,df.shape)
```

##### Verified the number of rows and columns in each dataset.

# Step 5: Check Dataset Information

```python
for name,df in datasets.items():
    print(name)
    df.info()
```

##### Reviewed column names, data types, and non-null values.

# Step 6: Check Missing Values

```python
for name,df in datasets.items():
    print(name)
    print(df.isnull().sum())
```

##### Identified missing values across all datasets.

# Step 7: Check Duplicate Records

```python
for name,df in datasets.items():
    print(name,df.duplicated().sum())
```

##### Checked all datasets for duplicate records.

# Step 8: Remove Duplicate Records

```python
for df in datasets.values():
    df.drop_duplicates(inplace=True)
```

##### Removed duplicate records to improve data quality.

# Step 9: Convert Date Columns

```python
covid_data["Date"]=pd.to_datetime(covid_data["Date"])
country_daywise["Date"]=pd.to_datetime(country_daywise["Date"])
daywise["Date"]=pd.to_datetime(daywise["Date"])
```

##### Converted date columns into datetime format for time-series analysis.

# Step 10: Rename Columns

```python
vaccination.rename(columns={"Doses administered per 100 people":"doses_administered_per_100_people","Total doses administered":"total_doses_administered","% of population vaccinated":"population_vaccinated_percent","% of population fully vaccinated":"population_fully_vaccinated_percent"},inplace=True)
```

##### Renamed columns using a consistent snake_case naming convention.

# Step 20: Feature Engineering

```python
countrywise["Recovery_Percentage"]=(countrywise["Recovered"]/countrywise["Confirmed"])*100
countrywise["Death_Percentage"]=(countrywise["Deaths"]/countrywise["Confirmed"])*100
```

##### Created new percentage-based features for deeper analysis.

# Step 21: Merge Datasets

```python
merged=pd.merge(countrywise,vaccination,on="Country",how="left")
```

##### Merged COVID-19 statistics with vaccination data using the Country column.

# Step 22: Export Clean Dataset

```python
# Step 22: Export Cleaned Datasets to Excel

output_file = "GLOBAL_COVID19_Cleaned_Data.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    covid_data.to_excel(writer, sheet_name="covid_19_data", index=False)
    country_daywise.to_excel(writer, sheet_name="country_daywise", index=False)
    countrywise.to_excel(writer, sheet_name="countrywise", index=False)
    country.to_excel(writer, sheet_name="Country", index=False)
    vaccination.to_excel(writer, sheet_name="Vaccination", index=False)
    vaccination_detail.to_excel(writer, sheet_name="Vaccinations_detail", index=False)
    daywise.to_excel(writer, sheet_name="daywise", index=False)

print(f"All cleaned datasets have been exported successfully to '{output_file}'.")
```

##### Exported the cleaned dataset for SQL analysis and Power BI dashboard development.
